**Context:**

The dataset comprises two attributes: "Hours" (independent) and "Scores" (dependent).

**Objective:**

To forecast the student's percentage based on their study hours.

**Prediction:**

What would be the anticipated score if a student studies for 9.25 hours per day?

**Data Source:**

Access the dataset via the following link: http://bit.ly/w-data

```python
#import important libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python

# load the data
df = pd.read_csv("https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
```

```python

# Visualize the relationship between study hours and scores
plt.figure(figsize=(8, 6))
plt.scatter(df['Hours'], df['Scores'], color='blue')
plt.title('Study Hours vs Scores')
plt.xlabel('Study Hours')
plt.ylabel('Scores')
plt.grid(True)
plt.show()
```

```python

# Prepare the data
X = df['Hours'].values.reshape(-1, 1)  # independent variable
y = df['Scores'].values  # dependent variable
```

```python

# Create and fit the model
model = LinearRegression()
model.fit(X, y)
```

```python

# Visualize the regression line
plt.figure(figsize=(8, 6))
plt.scatter(df['Hours'], df['Scores'], color='blue')
plt.plot(X, model.predict(X), color='red')
plt.title('Study Hours vs Scores (Linear Regression)')
plt.xlabel('Study Hours')
plt.ylabel('Scores')
plt.grid(True)
plt.show()
```

```python

# Predict the score for 9.25 hours/day
hours = 9.25
predicted_score = model.predict([[hours]])
predicted_score[0]
```
